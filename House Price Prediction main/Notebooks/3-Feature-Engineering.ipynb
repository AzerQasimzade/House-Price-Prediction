```python
# Import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats

# Display all columns
pd.set_option("display.max_columns", None)

print("Libraries imported successfully!")

# Load data

X = pd.read_csv("../Outputs/pipeline1_train_preprocessed.csv")
y = pd.read_csv("../Outputs/pipeline1_train_target.csv")["SalePrice"]

# Original cleaned dataset
df_original = pd.read_csv("../Outputs/pipeline1_clean.csv")

# Working copy for feature engineering
df_fe = df_original.copy()

print("X shape:", X.shape)
print("y shape:", y.shape)
print("Original data shape:", df_original.shape)

print("\nData loaded successfully.")
```

```python
print("Starting number of features:", df_fe.shape[1])
# Creating Total Square Footage feature

df_fe["TotalSF"] = (
    df_fe["TotalBsmtSF"]
    + df_fe["1stFlrSF"]
    + df_fe["2ndFlrSF"]
)

print("TotalSF feature created successfully.")

# Preview
df_fe[["TotalBsmtSF", "1stFlrSF", "2ndFlrSF", "TotalSF"]].head()
```

```python
# I Created Total Bathrooms feature

df_fe["TotalBathrooms"] = (
    df_fe["FullBath"]
    + 0.5 * df_fe["HalfBath"]
    + df_fe["BsmtFullBath"]
    + 0.5 * df_fe["BsmtHalfBath"]
)

print("TotalBathrooms feature created successfully.")

# Preview
df_fe[
    [
        "FullBath",
        "HalfBath",
        "BsmtFullBath",
        "BsmtHalfBath",
        "TotalBathrooms",
    ]
].head()
```

```python
# Create Total Porch Area feature

df_fe["TotalPorchSF"] = (
    df_fe["OpenPorchSF"]
    + df_fe["EnclosedPorch"]
    + df_fe["3SsnPorch"]
    + df_fe["ScreenPorch"]
)

print("TotalPorchSF feature created successfully.")

df_fe[
    [
        "OpenPorchSF",
        "EnclosedPorch",
        "3SsnPorch",
        "ScreenPorch",
        "TotalPorchSF",
    ]
].head()


# Create Total Rooms feature
#maybe
df_fe["TotalRooms"] = (
    df_fe["TotRmsAbvGrd"]
    + df_fe["KitchenAbvGr"]
)

print("TotalRooms feature created successfully.")


df_fe[
    [
        "TotRmsAbvGrd",
        "KitchenAbvGr",
        "TotalRooms",
    ]
].head()


# Create Total Outdoor Area feature

df_fe["TotalOutdoorArea"] = (
    df_fe["WoodDeckSF"]
    + df_fe["OpenPorchSF"]
    + df_fe["EnclosedPorch"]
    + df_fe["3SsnPorch"]
    + df_fe["ScreenPorch"]
)

print("TotalOutdoorArea feature created successfully.")

df_fe[
    [
        "WoodDeckSF",
        "OpenPorchSF",
        "EnclosedPorch",
        "3SsnPorch",
        "ScreenPorch",
        "TotalOutdoorArea",
    ]
].head()
```

```python
# Living area to lot area ratio

df_fe["LivingArea_to_LotArea"] = (
    df_fe["GrLivArea"] / df_fe["LotArea"]
)

print("LivingArea_to_LotArea feature created successfully.")

df_fe[
    [
        "GrLivArea",
        "LotArea",
        "LivingArea_to_LotArea",
    ]
].head()



#experiment
# df_fe["BasementRatio"] = (
#     df_fe["TotalBsmtSF"] / df_fe["TotalSF"]
# )

#experiment 
# df_fe["BasementRatio"] = (
#     df_fe["TotalBsmtSF"] / df_fe["TotalSF"]
# )
```

```python
# Created House Age feature

df_fe["HouseAge"] = df_fe["YrSold"] - df_fe["YearBuilt"]

print("HouseAge feature created successfully.")

df_fe[
    [
        "YrSold",
        "YearBuilt",
        "HouseAge",
    ]
].head()


# Created Remodel Age feature

df_fe["RemodelAge"] = df_fe["YrSold"] - df_fe["YearRemodAdd"]

print("RemodelAge feature created successfully.")

df_fe[
    [
        "YrSold",
        "YearRemodAdd",
        "RemodelAge",
    ]
].head()


# Created IsRemodeled feature

df_fe["IsRemodeled"] = (
    df_fe["YearBuilt"] != df_fe["YearRemodAdd"]
).astype(int)

print("IsRemodeled feature created successfully.")

df_fe[
    [
        "YearBuilt",
        "YearRemodAdd",
        "IsRemodeled",
    ]
].head()

#maybe
# df_fe["IsNewHouse"] = (df_fe["HouseAge"] <= 5).astype(int)
```

```python
# Create boolean features

df_fe["HasSecondFloor"] = (df_fe["2ndFlrSF"] > 0).astype(int)

df_fe["HasBasement"] = (df_fe["TotalBsmtSF"] > 0).astype(int)

df_fe["HasGarage"] = (df_fe["GarageArea"] > 0).astype(int)

df_fe["HasFireplace"] = (df_fe["Fireplaces"] > 0).astype(int)

df_fe["HasPool"] = (df_fe["PoolArea"] > 0).astype(int)

print("Boolean features created successfully.")

df_fe[
    [
        "HasSecondFloor",
        "HasBasement",
        "HasGarage",
        "HasFireplace",
        "HasPool",
    ]
].head()
```

```python
# Overall quality × living area
df_fe["Quality_x_Area"] = (
    df_fe["OverallQual"] * df_fe["GrLivArea"]
)

# Overall quality × total square footage
df_fe["Quality_x_TotalSF"] = (
    df_fe["OverallQual"] * df_fe["TotalSF"]
)

print("Interaction features created successfully.")

df_fe[
    [
        "OverallQual",
        "GrLivArea",
        "TotalSF",
        "Quality_x_Area",
        "Quality_x_TotalSF",
    ]
].head()
```

```python
# Identify newly created features

new_features = [
    col for col in df_fe.columns
    if col not in df_original.columns
]

print("Original features :", len(df_original.columns))
print("New features      :", len(new_features))
print("Total features    :", len(df_fe.columns))

# List all new features

print("\nNew Features Created:\n")

for i, feature in enumerate(new_features, start=1):
    print(f"{i}. {feature}")


# Summary statistics of engineered features

df_fe[new_features].describe()
```

```python
# Load cleaned test dataset

test_df = pd.read_csv("../Outputs/pipeline1_test_clean.csv")

# Working copy for feature engineering
test_fe = test_df.copy()

print("Test data shape:", test_fe.shape)



# Aggregation Features


# Total Square Footage
test_fe["TotalSF"] = (
    test_fe["TotalBsmtSF"]
    + test_fe["1stFlrSF"]
    + test_fe["2ndFlrSF"]
)

# Total Bathrooms
test_fe["TotalBathrooms"] = (
    test_fe["FullBath"]
    + 0.5 * test_fe["HalfBath"]
    + test_fe["BsmtFullBath"]
    + 0.5 * test_fe["BsmtHalfBath"]
)

# Total Porch Area
test_fe["TotalPorchSF"] = (
    test_fe["OpenPorchSF"]
    + test_fe["EnclosedPorch"]
    + test_fe["3SsnPorch"]
    + test_fe["ScreenPorch"]
)

# Total Rooms
test_fe["TotalRooms"] = (
    test_fe["TotRmsAbvGrd"]
    + test_fe["KitchenAbvGr"]
)

# Total Outdoor Area
test_fe["TotalOutdoorArea"] = (
    test_fe["WoodDeckSF"]
    + test_fe["OpenPorchSF"]
    + test_fe["EnclosedPorch"]
    + test_fe["3SsnPorch"]
    + test_fe["ScreenPorch"]
)



# Ratio Features


# Living Area to Lot Area
test_fe["LivingArea_to_LotArea"] = (
    test_fe["GrLivArea"] / test_fe["LotArea"]
)



# Temporal Features


# House Age
test_fe["HouseAge"] = (
    test_fe["YrSold"] - test_fe["YearBuilt"]
)

# Remodel Age
test_fe["RemodelAge"] = (
    test_fe["YrSold"] - test_fe["YearRemodAdd"]
)

# Is Remodeled
test_fe["IsRemodeled"] = (
    test_fe["YearBuilt"] != test_fe["YearRemodAdd"]
).astype(int)



# Boolean Features


test_fe["HasSecondFloor"] = (
    test_fe["2ndFlrSF"] > 0
).astype(int)

test_fe["HasBasement"] = (
    test_fe["TotalBsmtSF"] > 0
).astype(int)

test_fe["HasGarage"] = (
    test_fe["GarageArea"] > 0
).astype(int)

test_fe["HasFireplace"] = (
    test_fe["Fireplaces"] > 0
).astype(int)

test_fe["HasPool"] = (
    test_fe["PoolArea"] > 0
).astype(int)



# Interaction Features


# Overall Quality × Living Area
test_fe["Quality_x_Area"] = (
    test_fe["OverallQual"] * test_fe["GrLivArea"]
)

# Overall Quality × Total Square Footage
test_fe["Quality_x_TotalSF"] = (
    test_fe["OverallQual"] * test_fe["TotalSF"]
)



# Summary


test_new_features = [
    col for col in test_fe.columns
    if col not in test_df.columns
]

print("Original test features :", len(test_df.columns))
print("New features created   :", len(test_new_features))
print("Total test features    :", len(test_fe.columns))

print("\nNew Features Applied:\n")

for i, feature in enumerate(test_new_features, start=1):
    print(f"{i}. {feature}")

print("\nFeature Engineering applied to test dataset successfully.")







```
